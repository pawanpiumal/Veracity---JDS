```python
import pandas as pd  
import json
```

```python
df = pd.DataFrame()
```

```python
start = 1
num = 10000
```

```python
f = open('./dev.jsonl',encoding='utf-8')
lines = f.readlines()
```

```python
if(len(lines)<num):
    num = len(lines)
```

```python
for rowNumber in range(start-1,start-1+num):
    item = json.loads(lines[rowNumber])
    array = []
    for e in item['annotations']:
        if(e['short_answers'] != []):
            shortArray = []
            for token in e['short_answers']:
                stringArr = []
                for i in range(token['start_token'],token['end_token']):
                    stringArr.append(item['document_tokens'][i]['token'])
                shortArray.append(" ".join(stringArr))
            array.append(shortArray)
    newItem = {
        'index':rowNumber+1,
        "document_title":item['document_title'],
        'question_text':item['question_text'],
        'yes_no_answer':list(map(lambda x: x['yes_no_answer'],item['annotations']))
    }
    for no in range(0,len(array)):
        newItem[f'short_answer{no+1}']= array[no]
    df2 = pd.DataFrame([newItem])
    df = pd.concat([df, df2])
    # print(newItem)
    
```

```python
df
```

```python
df.to_csv(f'file{start}-{start+num-1}.csv')
```
